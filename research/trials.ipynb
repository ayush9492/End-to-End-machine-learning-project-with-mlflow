```python
def get_product(x: int, y: int) -> int:
    return x*y
```

```python
get_product(2, 3)
```

```python
get_product(2, "3")
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x: int, y: int) -> int:
    return x*y
```

```python
get_product(2, "3") # now it throws an error because of the type annotations and the ensure_annotations decorator.
# as we passed a string instead of an integer, it will raise an error.
```
